## Morphological Transformations


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
im = cv2.imread("j-images.png")
plt.imshow(im),plt.title("Original image")
plt.show()
```

```python
# Erosion
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(im, kernel, iterations = 1)

plt.imshow(erosion),plt.title("Erosion")
plt.show()
```

```python
# Erosion iterations = 2
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(im, kernel, iterations = 2)

plt.imshow(erosion),plt.title("Erosion")
plt.show()
```

```python
# Dilation
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(im, kernel, iterations = 1)

plt.imshow(dilation),plt.title("Dilation")
plt.show()
```

```python
# Erosion > Dilation
kernel = np.ones((5,5),np.uint8)
dilation2 = cv2.dilate(erosion, kernel, iterations = 1)

plt.imshow(dilation2),plt.title("Erosion > Dilation")
plt.show()
```

```python
# Opening
kernel = np.ones((5,5),np.uint8)
opening = cv2.morphologyEx(im, cv2.MORPH_OPEN, kernel)

plt.imshow(opening),plt.title("Opening")
plt.show()
```

```python
# Closing
kernel = np.ones((5,5),np.uint8)
closing = cv2.morphologyEx(im, cv2.MORPH_CLOSE, kernel)

plt.imshow(opening),plt.title("Closing")
plt.show()
```

### Morphological Gradient

```python
# Dilation - Erosion
kernel = np.ones((5,5),np.uint8)
gradient = cv2.morphologyEx(im, cv2.MORPH_GRADIENT, kernel)

plt.imshow(gradient),plt.title("Morphological Gradient")
plt.show()
```

```python
# Top Hat

kernel = np.ones((5,5),np.uint8)
topHat = cv2.morphologyEx(im, cv2.MORPH_TOPHAT, kernel)

plt.imshow(topHat),plt.title("Top Hat")
plt.show()
```

```python
# Black Hat

kernel = np.ones((5,5),np.uint8)
blackHat = cv2.morphologyEx(im, cv2.MORPH_TOPHAT, kernel)

plt.imshow(blackHat),plt.title("Black Hat")
plt.show()
```
