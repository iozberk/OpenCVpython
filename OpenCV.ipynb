```python
# OpenCv Rectangle
im = cv2.imread("OpenCV.png")

im2 = cv2.rectangle(im,(290,5),(900,525),[0,0,255],9)
im2 = cv2.rectangle(im,(5,540),(560,1120),[0,0,255],9)
im2 = cv2.rectangle(im,(600,540),(1195,1120),[0,0,255],9)

#cv2.imshow("OpenCV Logo Rectangle: ", im2)
#cv2.waitKey(0)
#cv2.destroyAllWindows()

plt.imshow(im2)
plt.show()
```

```python
# images size down
im = cv2.imread("OpenCV.png")


smallIm = cv2.pyrDown(im) # cv2.pyrUp 2 times bigger
plt.imshow(smallIm)
plt.show()
print(im.shape) # Normal Shape (739, 600, 3)
print(smallIm.shape) # Small Shape (739, 600, 3) 2 times smaller
```

```python
# Live Video Capture

camera = cv2.VideoCapture(0)

while True:
    ret,video = camera.read()
    cv2.imshow("Live Video: ",video)
    if cv2.waitKey(24) & 0xFF ==' q ':
        break
camera.release()
cv2.destroyAllWindows()
```

```python
# Flipping an image
# flipping the image across the x-axis, the y-axis and then across both axes
import cv2
im = cv2.imread("OpenCV.png")
flipV = cv2.flip(im,0) # Vertical
flipH = cv2.flip(im,1) # Horizontal
flipX = cv2.flip(im,1) # Horizontal
plt.imshow(flipV)
plt.show()
```

```python
# Concatenating images
import numpy as np
import cv2 
import matplotlib.pyplot as plt

im = cv2.imread("OpenCV.png")
a = np.vstack((im,im))            # Vertical
b = np.hstack((im,im,im))         # Horizontal

plt.imshow(a) # Vertical
plt.show()
```

```python
# Concatenating images BGR - Gray
import numpy as np
import cv2 
import matplotlib.pyplot as plt

im = cv2.imread("OpenCV.png")
imGray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
gray = cv2.cvtColor(imGray,cv2.COLOR_GRAY2BGR)

con = np.hstack((im,gray))

plt.imshow(con) # Vertical
plt.show()
```

```python
# Converting an image to black and white
import numpy as np
import cv2 
import matplotlib.pyplot as plt

im2 = cv2.imread("OpenCV.png")
grayim = cv2.cvtColor(im2, cv2.COLOR_BGR2GRAY)
(thresh, baw) = cv2.threshold(grayim, 127, 255, cv2.THRESH_BINARY)
 
cv2.imshow("Black and White",baw)
 
cv2.waitKey(0)
cv2.destroyAllWindows()
    
#plt.imshow(baw) 
#plt.show()
```
