```python
import cv2
import matplotlib.pyplot as plt
im = cv2.imread("OpenCV.jpg")
plt.imshow(im)
plt.show()
```

```python
mean3 = cv2.blur(im,(3,3))
plt.imshow(mean3),plt.title('3x3 Filtering')
plt.show()
```

```python
mean7 = cv2.blur(im,(7,7))
plt.imshow(mean7),plt.title('7x7 Filtering')
plt.show()
```

```python
median3 = cv2.medianBlur(im,3)
plt.imshow(median3),plt.title('Median 3')
plt.show()
```

```python
median7 = cv2.medianBlur(im,7)
plt.imshow(median7),plt.title('Median 7')
plt.show()
```

```python
gauss3 = cv2.GaussianBlur(im,(3,3),0)
plt.imshow(gauss3),plt.title('Gauss 3')
plt.show()
```

```python
gauss7 = cv2.GaussianBlur(im,(7,7),0)
plt.imshow(gauss7),plt.title('Gauss 7')
plt.show()
```
